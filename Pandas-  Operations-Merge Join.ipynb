```python
#Merge Operation
```

```python
data4 = {'emp_id' :[101,102,103,104],
         'salary' : [3456,6533,6754,7865],
         'pf' : [23,45,67,45]
    
}
```

```python
data5 = {'emp_id' :[101,102,104,105],
         'mobile_no' : [73456,76533,76754,77865],
         'house_no' : [123,145,167,145]
    
}
```

```python
df4 = pd.DataFrame(data4)
```

```python
df5 = pd.DataFrame(data5)
```

```python
df4
```

```python
df5
```

```python
pd.merge(df4,df5)  # takes all common values
```

```python
pd.merge(df4,df5, how = 'left') # left(df4)& right(df5)-- it will consider whatever we have in left dataframe(df4)
```

```python
pd.merge(df4,df5, how = 'right')
```

```python
pd.merge(df4,df5, how = 'cross')
```

```python
pd.merge(df4,df5, how = 'right', on ='emp_id')
```

```python
data6 = {'emp_id1' :[101,102,103,104],
         'salary' : [3456,6533,6754,7865],
         'pf' : [23,45,67,45]
    
}

data7 = {'emp_id2' :[101,102,104,105],
         'mobile_no' : [73456,76533,76754,77865],
         'house_no' : [123,145,167,145]
    
}
```

```python
df6 = pd.DataFrame(data6)
```

```python
df7 = pd.DataFrame(data7)
```

```python
pd.merge(df6,df7) # no common column
```

```python
pd.merge(df6,df7, left_on ='emp_id1', right_on = 'emp_id2') # If we don't have common column then we can use Join operation
```

```python
df_sales = pd.read_csv(r'C:\Users\Lenovo\Documents\Data Science\Data Science iNeuron\Python\2023\Pandas\Data Set 2\sales_data_final.csv')
```

```python
df_sales. head()
```

```python
def profit_flag(a):
    if a >= 0:
        return 'Positive'
    else:
        return 'Negative'
```

```python
profit_flag(56)
```

```python
df_sales['flag_profit'] = df_sales['profit'].apply(profit_flag)
```

```python
df_sales.head()
```

```python
df_sales['len_customer_name'] = df_sales['customer_name'].apply(len)
```

```python
df_sales.head()
```

```python
def quantity_flag(a):
    if a <= 10 :
        return 'Low'
    elif a >10  and a <20 :
        return 'Medium'
    else:
        return 'high'
```

```python
df_sales['flag_quantity'] = df_sales['quantity'].apply(quantity_flag)
```

```python
df_sales.head(20)
```

```python
 df_sales['square_quantity'] =  df_sales['quantity'].apply(lambda a:a**2)
```
