```python
dfal['Date_Month'] = pd.to_datetime(dfap['Date']).dt.month
```

```python
dfal['Duration_Hours'] = pd.to_timedelta(dfal['Duration']).dt.total_seconds()/3600 
# Calculate the duration in hours
```

```python
dfal.head(2)
```

```python
agent_Hrs_weekly= dfal.groupby(['Agent','Date_Week'])['Duration_Hours'].sum().reset_index()
```

```python
agent_Hrs_weekly
```

```python
# Group the data by 'Agent' and 'Date_Month', and calculate the total active hours for each group
total_active_hours = dfal.groupby(['Agent', 'Date_Month'])['Duration_Hours'].sum().reset_index()
```

```python
total_active_hours
```

```python
# Calculate the total number of hours in the month
total_hours_in_month = dfal.groupby('Date_Month').size().reset_index(name='Total_Hours')
```

```python
total_hours_in_month
```

```python
total_hours_in_month['Total_Hours'] = total_hours_in_month['Total_Hours'] * 24
```

```python
merged_data = total_active_hours.merge(total_hours_in_month, on='Date_Month')
```

```python
# Calculate the percentage of active hours for each month
merged_data['Percentage_Active_Hours'] = (merged_data['Duration_Hours'] / merged_data['Total_Hours']) * 100
```
